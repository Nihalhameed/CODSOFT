```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
df
```

```python
df.rename(columns={'v1':'Category','v2':'Message'},inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Category']=lab.fit_transform(df['Category'])
df
```

```python
data=df.copy()
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
#Cleaning the text :
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
lst=[]
for i in range(0,len(data)):
  messages=re.sub('[^a-zA-Z0-9]+',' ',data['Message'][i]) #removing special characters from sentences
  messages=messages.lower() #lowering the sentences
  messages=messages.split() #tokenization : sentences to word by word data
  messages=[lemma.lemmatize(word) for word in messages if word not in stopwords.words('english')] #remove stopwords and perform lemmatization
  messages=' '.join(messages)
  lst.append(messages)
```

```python
lst[0]
```

```python
#Using of TF-Id Vectorizer to convert into numerical data:
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_v=TfidfVectorizer(max_features=5000)
train_data=tfidf_v.fit_transform(lst).toarray()
```

```python
train_data.shape
```

```python
y=df['Category'].values
y
```

```python
#Training and testing data :
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
naive_model=MultinomialNB()
svm_model=SVC()
neighbours_model=KNeighborsClassifier(n_neighbors=5)
lst_models=[neighbours_model,naive_model,svm_model]
```

```python
for i in lst_models:
  print("Model name............. : ",i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("**********************************")
  print("Accuracy: ")
  print(accuracy_score(y_test,y_pred))
  print(confusion_matrix(y_test,y_pred))
  print(classification_report(y_test,y_pred))
```
